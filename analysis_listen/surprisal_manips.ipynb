```python
import matplotlib.pyplot as plt 
from scipy.io import loadmat, savemat
import pickle
```

```python
surprisal = loadmat('AM_concat_2.mat')
```

```python
surprisal_toplot = surprisal['out']['surprisal'][0][0]


plt.figure(figsize = (20,8))
plt.plot(surprisal_toplot[0:328], marker = 'o')
plt.ylim(0,25)
```

```python
# Number of subplots and values per subplot
n_subplots = 4
values_per_subplot = 328

# Create a figure with 4 subplots
plt.figure(figsize=(20, 15))

# Loop to create each subplot
for i in range(n_subplots):
    start_index = i * values_per_subplot
    end_index = start_index + values_per_subplot
    plt.subplot(n_subplots, 1, i + 1)  # Create a subplot in a 4-row grid
    plt.plot(surprisal_toplot[start_index:end_index], marker='o')
    plt.ylim(8, 25)
    plt.xlim(0,328)
    plt.xlabel('Index')  # Add x-axis label
    plt.ylabel('Surprisal Value')  # Add y-axis label

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
