### Set up 

```python
#-----------------------------
#           PARAMETERS
#-----------------------------
evoked_dir = '/Users/cindyzhang/Documents/M2/Audiomotor_Piano/AM-EEG/analysis_motor/motor_ERP_data'


period = 'post'
subjects_to_process = ['01', '05']

```

#### TO DO: take equal subsets from each participant?

```python
#-----------------------------
#           GROUPING EVOKEDS
#-----------------------------

evoked_list = []

for file in sorted(os.listdir(evoked_dir)):
    file = glob.glob(os.path.join(evoked_dir, f'motor_ERP_{period}_??.fif'))[0]
    evoked = mne.read_evokeds(file)[0]
    evoked_list.append(evoked)


concatenated_evoked = mne.combine_evoked(evoked_list, weights = 'equal')
```

### Plot grand average

```python
concatenated_evoked.plot_joint(times =[ 0, 0.12, 0.2], title= f'Motor-only keystrokes {period}-training (grand average)')
```
